## Project Description: Next Word Prediction Using LSTM
#### Project Overview:

This project aims to develop a deep learning model for predicting the next word in a given sequence of words. The model is built using Long Short-Term Memory (LSTM) networks, which are well-suited for sequence prediction tasks. The project includes the following steps:

1- Data Collection: We use the text of Shakespeare's "Hamlet" as our dataset. This rich, complex text provides a good challenge for our model.

2- Data Preprocessing: The text data is tokenized, converted into sequences, and padded to ensure uniform input lengths. The sequences are then split into training and testing sets.

3- Model Building: An LSTM model is constructed with an embedding layer, two LSTM layers, and a dense output layer with a softmax activation function to predict the probability of the next word.

4- Model Training: The model is trained using the prepared sequences, with early stopping implemented to prevent overfitting. Early stopping monitors the validation loss and stops training when the loss stops improving.

5- Model Evaluation: The model is evaluated using a set of example sentences to test its ability to predict the next word accurately.

6- Deployment: A Streamlit web application is developed to allow users to input a sequence of words and get the predicted next word in real-time.

```python
# Data Collection
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
import pandas as pd

# Load the dataset for Hamlet, Macbeth and Julius Caesar from the Gutenberg corpus
hamlet_data = gutenberg.raw('shakespeare-hamlet.txt')
macbeth_data = gutenberg.raw('shakespeare-macbeth.txt')
caesar_data = gutenberg.raw('shakespeare-caesar.txt')

# Save the Hamlet text to a file
with open('hamlet.txt', 'w') as file:
    file.write(hamlet_data)

# Save the Macbeth and Caesar texts to their respective files
with open('macbeth.txt', 'w') as file:
    file.write(macbeth_data)

with open('caesar.txt', 'w') as file:
    file.write(caesar_data)

# Combine the text content of all three works
final_text = hamlet_data + "\n" + macbeth_data + "\n" + caesar_data

# Save the combined text into 'final.txt'
with open('final.txt', 'w') as file:
    file.write(final_text)
```

```python
## Data Preprocessing

import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

##laod the dataset
with open('final.txt','r') as file:
    text=file.read().lower()

## Tokenize the text-creating indexes for words
tokenizer=Tokenizer()
tokenizer.fit_on_texts([text])
total_words=len(tokenizer.word_index)+1
total_words
```

```python
tokenizer.word_index
```

```python
## create input sequences
input_sequences=[]
for line in text.split('\n'):
    token_list=tokenizer.texts_to_sequences([line])[0]
    for i in range(1,len(token_list)):
        n_gram_sequence=token_list[:i+1]
        input_sequences.append(n_gram_sequence)

```

```python
input_sequences
```

```python
# Pad Sequences
max_sequence_len=max([len(x) for x in input_sequences])
max_sequence_len
```

```python
input_sequences=np.array(pad_sequences(input_sequences,maxlen=max_sequence_len,padding='pre'))
input_sequences
```

```python
##create predicitors and label
import tensorflow as tf
x,y=input_sequences[:,:-1],input_sequences[:,-1]
```

```python
x
```

```python
y
```

```python
y=tf.keras.utils.to_categorical(y,num_classes=total_words)
y
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Define early stopping
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
```

```python
## Train our LSTM RNN

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout,GRU

## Define the model
model=Sequential()
model.add(Embedding(total_words,100,input_length=max_sequence_len-1))
model.add(LSTM(150,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words,activation="softmax"))

# #Compile the model
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
## GRU RNN
## Define the model
model=Sequential()
model.add(Embedding(total_words,100,input_length=max_sequence_len-1))
model.add(GRU(150,return_sequences=True))
model.add(Dropout(0.2))
model.add(GRU(100))
model.add(Dense(total_words,activation="softmax"))

# #Compile the model
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
## Train the model
history=model.fit(x_train,y_train,epochs=50,validation_data=(x_test,y_test),verbose=1,callbacks=[early_stopping])
```

```python
# Function to predict the next word
def predict_next_word(model, tokenizer, text, max_sequence_len):
    token_list = tokenizer.texts_to_sequences([text])[0]
    if len(token_list) >= max_sequence_len:
        token_list = token_list[-(max_sequence_len-1):]  # Ensure the sequence length matches max_sequence_len-1
    token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
    predicted = model.predict(token_list, verbose=0)
    predicted_word_index = np.argmax(predicted, axis=1)
    for word, index in tokenizer.word_index.items():
        if index == predicted_word_index:
            return word
    return None
```

```python
input_text="To be or not to be"
print(f"Input text:{input_text}")
max_sequence_len=model.input_shape[1]+1
next_word=predict_next_word(model,tokenizer,input_text,max_sequence_len)
print(f"Next Word PRediction:{next_word}")
```

```python
def predict_next_word(model, tokenizer, input_text, max_sequence_len):
    # Convert input text to sequence of integers
    sequence = tokenizer.texts_to_sequences([input_text])
    sequence = pad_sequences(sequence, maxlen=max_sequence_len, padding='pre')

    # Predict the next word index
    predicted_probabilities = model.predict(sequence)
    predicted_index = np.argmax(predicted_probabilities, axis=-1)

    # Get the word corresponding to the predicted index
    predicted_word = tokenizer.index_word[predicted_index[0]]
    return predicted_word

# Define the input text and the number of predictions
input_text = "To be or not to be"
print(f"Input text: {input_text}")

# Set max sequence length based on the model's input shape
max_sequence_len = model.input_shape[1] + 1

# Perform the prediction 10 times
for _ in range(5):
    next_word = predict_next_word(model, tokenizer, input_text, max_sequence_len)
    input_text += ' ' + next_word  # Append the predicted word to the input text
    print(f"Next Word Prediction: {next_word}")
```
